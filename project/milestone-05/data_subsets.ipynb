```python
# Import the tables of the data set as dataframes.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.api.types import is_numeric_dtype


DATA_DIR = '../milestone-04/data/'

users = pd.read_csv('{}/calcularis_small_users.csv'.format(DATA_DIR))
events = pd.read_csv('{}/calcularis_small_events.csv'.format(DATA_DIR))
subtasks = pd.read_csv('{}/calcularis_small_subtasks.csv'.format(DATA_DIR))
```

```python
sns.set_palette('viridis', n_colors=4)
```

```python
users_per_country = users.country.value_counts()
```

```python
users_per_country
```

```python
def my_autopct(pct):
    return ('%.0f' % pct) +'%' if pct > 1 else ''
```

```python
labels = list(users_per_country.index[:4])
labels.extend(([""] * 4))
labels.extend(["Other"])
labels.extend(([""] * 4))
```

```python
len(labels)
```

```python
plt.pie(users_per_country, labels=labels, autopct=my_autopct)
plt.title("Random sample")
plt.suptitle("Percent of users by country")
```
